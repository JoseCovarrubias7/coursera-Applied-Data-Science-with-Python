```python
import pandas as pd

#Adding our CVS files
df1 = pd.read_csv('assets/2019-2020_Liga_MX_season_.csv')
df2 = pd.read_csv('assets/2020-2021_Liga_MX_season_.csv')
df3 = pd.read_csv('assets/2021-2022_Liga_MX_season_.csv')
df4 = pd.read_csv('assets/2022-2023_Liga_MX_season_.csv')

# Rename the "Teamvte" column to "Team" in the last three files.
df2 = df2.rename(columns={'Teamvte': 'Team'})
df3 = df3.rename(columns={'Teamvte': 'Team'})
df4 = df4.rename(columns={'Teamvte': 'Team'})

# Concatenate the four files
df_concat = pd.concat([df1, df2, df3, df4])

#Rename Morelia as mazatlan, and eliminate veracruz since this team no longer exists.
df_concat.loc[df_concat['Team'] == 'Morelia', 'Team'] = 'Mazatlán'
df_concat = df_concat.drop(18)

# Apply replacement to all tables
df_concat['Team'] = df_concat['Team'].str.replace(r'\(C\)', '')


df_concat.head(100)
```

```python
import numpy as np

columns_to_drop = ['Qualification or relegation', 'Qualification']
df_concat['Team'] = df_concat['Team'].str.strip()

df_concat = df_concat.drop(columns=columns_to_drop)

df_concat['Pos'] = [i + 1 for i, _ in enumerate(df_concat['Pos'])]

# Define the conditions for assigning values to the "Temp" column.
condiciones = [
    (df_concat['Pos'] < 19),
    (df_concat['Pos'] >= 19) & (df_concat['Pos'] < 37),
    (df_concat['Pos'] >= 37) & (df_concat['Pos'] < 55),
    (df_concat['Pos'] >= 55) & (df_concat['Pos'] < 73)
]

# Define the values corresponding to the conditions
valores = ['2019-2020', '2020-2021', '2021-2022',"2022-2023"]

 # Add the new column "Temp"
df_concat['Temp'] = np.select(condiciones, valores, None)

groups = df_concat.groupby('Team')

dataframes = []

# Get the DataFrames of each group and add them to the list.
for name, group in groups:
    dataframes.append(group)

# Concatenate DataFrames into a single DataFrame
df_result = pd.concat(dataframes)

# Display the resulting DataFrame
df_result.head(50)

```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, Layout
import ipywidgets as widgets
import seaborn as sns

df_grouped = df_result.groupby('Team')
unique_teams = df_result['Team'].unique()
num_plots = len(unique_teams)

g = None  # Variable for storing the FacetGrid object

line_styles = {
    'W': {'linestyle': '-', 'linewidth': 2, 'color': 'green'},
    'D': {'linestyle': '-.', 'linewidth': 2, 'color': 'orange'},
    'L': {'linestyle': ':', 'linewidth': 2, 'color': 'red'}
}

legend_labels = {
    'W': 'Victories (W)',
    'D': 'Draws (D)',
    'L': 'Losses (L)'
}

def update_plot(selected_categories):
    global g  # Access the global variable g

    # Delete the previous graphic if it exists
    if g is not None:
        plt.close(g.fig)

    # Create a new FacetGrid object
    g = sns.FacetGrid(df_result, col='Team', hue='Team', col_wrap=6, height=3, legend_out=False)

    # Map lines and points according to the selected categories
    for category in selected_categories:
        g.map(plt.plot, 'Temp', category, **line_styles[category])
        g.map(plt.scatter, 'Temp', category, **line_styles[category], s=25)

    # Configure titles and captions
    g.set_titles("{col_name}")
    g.set_ylabels('Count') 
    plt.subplots_adjust(top=0.92)
    g.fig.suptitle('Data Analysis by Team for the last 4 seasons')
    handles = [plt.Line2D([], [], color=line_styles[label]['color'], linewidth=2) for label in selected_categories]
    labels = [legend_labels[label] for label in selected_categories]
    plt.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 4.15), ncol=3)

    # Rotate the legend on the X-axis
    for ax in g.axes.flat:
        plt.setp(ax.get_xticklabels(), rotation='vertical')

    # Show the graph
    plt.show()
    g.savefig('graphAssignment4.png', dpi=300, bbox_inches='tight', facecolor='white')

    
# Create options list and interactive widget
category_options = ['W', 'D', 'L']
interact(update_plot, selected_categories=widgets.SelectMultiple(options=category_options, value=category_options, description='Categories:'))
```
